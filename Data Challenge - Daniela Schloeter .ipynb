```python
def outliers(df, col):
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1 #Interquartile range
    fence_low  = q1-1.5*iqr
    fence_high = q3+1.5*iqr
    df_out = df.loc[(df[col] > fence_low) & (df[col] < fence_high)]
    return df_out
```

```python
def lineplot(dfCRD,dfCSH, col,time,week,number):
    dfCRD[col].plot(kind='line',label='CRD')
    plt.title(str(time)+' vs. Tip in credit card')
    plt.ylabel('Mean tip amount')
    plt.xlabel(str(time))
    plt.xticks(np.arange(number), week)
    plt.show()

    dfCSH[col].plot(kind='line',label='CSH')
    plt.title(str(time)+' vs. Tip in cash')
    plt.ylabel('Mean tip amount')
    plt.xlabel(str(time))
    plt.xticks(np.arange(number), week)
    plt.show()


    fig,ax=plt.subplots()

    dfCRD[col].plot(kind='line',label='CRD',ax=ax)
    dfCSH[col].plot(kind='line',label='CSH',ax=ax)
    ax.legend(loc='best')
    

    ax.set(title=str(time)+' vs.Tip. Credit and Cash',
       ylabel='Mean tip amount',
       xlabel=str(time))
    ax.legend(['CRD','CSH'])
    ax.set_xticks([])
    return plt.show()
```

# 2. Which driver (identified by hack license) had the largest change in tip amount from the first to the last week of June 2013?

For this part it is important to consider multiple details. First, we could use the first and last weeks of June starting on sunday or monday, or the first and last 7 days of June. I will use the by week option, setting the first day of the week as Monday. 
 
For June 2013 the first week starts on Monday 2 and ends on Sunday 8 inclusive. The last week of the month starts on Monday 23 and ends on Sunday 29 inclusive.

```python
first_7days=fare_June2013[fare_June2013.index.day.isin([2,3,4,5,6,7,8])]
last_7days=fare_June2013[fare_June2013.index.day.isin([23,24,25,26,27,28,29])]
```

```python
drivers_mean_first_week=first_7days.groupby(' hack_license').mean()
drivers_mean_last_week=last_7days.groupby(' hack_license').mean()
drivers_mean_first_week=drivers_mean_first_week.rename(columns = {' tip_amount':'first_tip_amount','tip_proportion':'first_tip_proportion'})
drivers_mean_last_week=drivers_mean_last_week.rename(columns = {' tip_amount':'last_tip_amount','tip_proportion':'last_tip_proportion'})
#drivers_mean_last_week
```

```python
drivers_mean_last_week=drivers_mean_last_week.loc[:,['last_tip_amount','last_tip_proportion']]
drivers_mean_first_week=drivers_mean_first_week.loc[:,['first_tip_amount','first_tip_proportion']]
difference=pd.concat([drivers_mean_last_week, drivers_mean_first_week], axis=1, sort=True)
```

```python
difference['difference_mean_tip_last_fist_days']=difference['last_tip_amount']-difference['first_tip_amount']
difference['difference_mean_tip_proportion_last_fist_days']=difference['last_tip_proportion']-difference['first_tip_proportion']
difference=difference.sort_values('difference_mean_tip_last_fist_days')
```

When taking a look at the 'difference' dataframe we have to consider that, some drivers may have not worked during one of the two periods(first/last week). We are going to calculate this proportion to find out how many drivers we are excluding from the analysis. 

Then, I will eliminate all the missing values from the concatenated dataframe 'difference'.

```python
#difference.describe()
difference=difference.dropna()
```

We can observe that out of 30,919 drivers in the first/last week of June 2013, 29,174 worked in both the first and the last week of June 2013, which is a good result because we did not loose much data.

However, it is important to consider that we can have both negative and positive changes.

I will make a top 5 list for the three scenarios. Positive, Negative, and General.

```python
# Negative

difference_Negative=difference['difference_mean_tip_last_fist_days']
print('These are the Drivers with the biggest Negative changes from the first to the last week of June 2013: ')
print()
print('Hack license Negative                  ','Difference mean tip (first and last week of June 2013) ')
print((difference_Negative.head()))

# Positive

difference=difference.sort_values('difference_mean_tip_last_fist_days',ascending=False)
difference_Positive=difference['difference_mean_tip_last_fist_days']
print()
print()
print('These are the Drivers with the biggest Positive changes from the first to the last week of June 2013: ')
print()
print('Hack license Positive                  ','Difference mean tip (first and last week of June 2013) ')
print((difference_Positive.head()))


# General

difference1=difference.abs()
difference1=difference1.sort_values('difference_mean_tip_last_fist_days',ascending=False)
difference1
difference_General=difference1['difference_mean_tip_last_fist_days']

print()
print()
print('These are the Drivers with the biggest changes either positive or negative from the first to last week of June 2013: ')
print()
print('Hack license General                ','Difference mean tip (first and last week of June 2013) ')
print((difference_General.head()))


# Negative Proportion

difference=difference.sort_values('difference_mean_tip_proportion_last_fist_days')
difference_Negative_Prop=difference['difference_mean_tip_proportion_last_fist_days']*100

print()
print()
print("But the change can also be in proportion. Lets calculate the changes for proportion's means")
print()
print()
print('These are the Drivers with the biggest Negative proportion changes from the first to the last week of June 2013: ')
print()
print('Hack license Negative Proportion        ','Difference mean proportion % (first and last week of June 2013) ')
print((difference_Negative_Prop.head()))


# Positive Proportion

difference=difference.sort_values('difference_mean_tip_proportion_last_fist_days',ascending=False)
difference_Positive_Prop=difference['difference_mean_tip_proportion_last_fist_days']*100
print()
print()
print('These are the Drivers with the biggest Positive proportion changes from the first to the last week of June 2013: ')
print()
print('Hack license Positive Proportion        ','Difference mean proportion % (first and last week of June 2013) ')
print((difference_Positive_Prop.head()))


# General Proportion

difference1=difference.abs()
difference1=difference1.sort_values('difference_mean_tip_proportion_last_fist_days',ascending=False)
difference1
difference_General_Prop=difference1['difference_mean_tip_proportion_last_fist_days']*100
print()
print()
print('These are the Drivers with the biggest changes either positive or negative proportion from the first to the last week of June 2013: ')
print()
print('Hack license General Proportion        ','Difference mean proportion % (first and last week of June 2013) ')
print((difference_General_Prop.head()))
```
